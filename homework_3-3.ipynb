# 깨지는 이유

화면에 출력하면 한글이 깨져있다. use_unicode=True설정을 주었는데도 그렇다. 다운로드 받으면서 한글이 깨져 있기 때문에 그렇다. 다운로드 받은 파일을 수정해서 출력하면 된다.

````python
import pyspark
import os

os.environ["PYSPARK_PYTHON"] = "/usr/bin/python3"
os.environ["PYSPARK_DRIVER_PYTHON"] = "/usr/bin/python3"

myConf=pyspark.SparkConf()
spark = pyspark.sql.SparkSession\
    .builder\
    .master("local")\
    .appName("myApp")\
    .config(conf=myConf)\
    .getOrCreate()

spark.version
````

````python
# l 누르면 라인수가 나온다.
# data bricks 쓸때 write 하지말고 드래그 해서 업로드해서 사용하라
````

````python
df1=spark.read.csv(os.path.join("data", "경기도 의정부시_인구현황_20210910.csv"))

print(f"type: {type(df1)}")

for i in df1.take(5):
    print(i)
````

````python
popRdd = spark.sparkContext\
    .textFile(os.path.join("data", "경기도 의정부시_인구현황_20210910.csv"))

print(f"type: {type(popRdd)}")

for i in popRdd.take(5):
    print(i)
````

````python
df2=spark.read.csv(os.path.join("data", "제주특별자치도 서귀포시_고령화비율및노령화지수현황_20210831.csv"))
for i in df2.take(5):
    print(i)
````

# 해결: 파일 고침 + utf8 저장

````python
fixdf1= spark.read.format("csv")\
    .option("delimiter", ",")\
    .option("ESCAPE quote", '"')\
    .option("header",True)\
    .load(os.path.join("data", "고침_경기도 의정부시_인구현황_20210910.csv"))
for i in fixdf1.take(5):
    print(i)
````

````python
fixdf2= spark.read.format("csv")\
    .option("delimiter", ",")\
    .option("ESCAPE quote", '"')\
    .option("header",True)\
    .load(os.path.join("data", "고침_제주특별자치도 서귀포시_고령화비율및노령화지수현황_20210831.csv"))
for i in fixdf2.take(5):
    print(i)
````
